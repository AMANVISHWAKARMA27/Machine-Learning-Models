```python
import pandas as pd
df = pd.read_csv("spam.csv")
df.head()
```

```python
# Grouping the data by category
df.groupby('Category').describe()
```

```python
# Converting the category column into numbers so that the computer can understand
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.25)
```

```python
# We have the Message column which contains text, and we have to convert them into numbers
# Now we will use CountVectorizer

from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

###CountVectorizer

**CountVectorizer** is a great tool provided by the scikit-learn library in Python. It is used to transform a given text into a vector on the basis of the frequency (count) of each word that occurs in the entire text. This is helpful when we have multiple such texts, and we wish to convert each word in each text into vectors (for using in further text analysis).

```python
# MultinomialNB
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
emails = [
    'Hey David, lets play together in the football ground.',
    'Sale sale sale, now you can get upto 70% off in the VBazar exclusive sale.'
]

emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

```python
# Creating a sklearn pipeline
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
clf.predict(emails)
```
